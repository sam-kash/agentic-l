### LANGCHAIN VERSION V1

```python
import os 
from dotenv import load_dotenv
load_dotenv()

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
from langchain.agents import create_agent

# 1. Create the agent using the new factory function.
agent = create_agent(
    model="gpt-4o", 
    tools=[],       
    system_prompt="act as a helpful assistant"
)

# 2. Run the agent using .invoke() and pass your prompt inside "messages"
response = agent.invoke({"messages": [("user", "Hello! How can you help me?")]})

# 3. Print the response
print(response["messages"][-1].content)
```
